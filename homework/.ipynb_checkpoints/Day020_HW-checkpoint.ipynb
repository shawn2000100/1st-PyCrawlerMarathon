```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
tr = table.find_all('tr')
d = {}
for row in tr[1:]:
    data = row.find_all('td')
    L = len(data)
    if L == 5:
        heading = data[0].text.strip()
        d[heading] = {}
        date, val = data[2].text.strip(), data[3].text.strip()
        d[heading][date] = val
    elif L == 3:
        date, val = data[0].text.strip(), data[1].text.strip()
        d[heading][date] = val
d
```
