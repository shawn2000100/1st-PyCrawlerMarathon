### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

articles = soup.find(class_='part_list_2')
article = articles.find_all('h3')

for d in article:
    print(d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

articles = soup.find(class_='part_list_2').find_all('h3')


val = {}
for d in articles:
    cat = d.find('em').text
    tit = d.find('a').text
    
    if cat not in val.keys():
        val[cat] = [tit]
    else:
        val[cat].append(tit)
        
val
```

### ③ 哪一個類別的文章最多

```python
for i in val:
    print(i, len(i))
```
